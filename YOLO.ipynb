```python
from ultralytics import YOLO
import cv2
import os
```

```python
#loading YOLO model
model=YOLO("yolo26n.pt")
results = model("/content/drive/MyDrive/Ai Workshop/hello.jpg")
img = results[0].plot()
output_path= os.path.join(os.getcwd(),"detected_output.jpg")
cv2.imwrite(output_path,img)
```
