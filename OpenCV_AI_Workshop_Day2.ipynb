```python
#Shapes in OpenCV
#1. Line  2. Circle  3. Rectangle  4. Polygon
```

```python
import cv2
```

```python
a=np.zeros([200,200,3])
cv2.line(a, (120, 20), (20, 160), (255, 0, 0), 3) #(img: MatLike, pt1: Point, pt2: Point, color: Scalar, thickness: int = ..., lineType: int = ..., shift: int = ...
cv2_imshow(a)
```

```python
cv2.circle(a,(100,100),100,(0,255,0),1, cv2.LINE_4)
cv2_imshow(a) # center: Point, radius: int, color: Scalar, thickness: int = ..., lineType: int = ..., shift: int = ...)
```

```python
cv2.rectangle(a,(30,30),(170,170),(200,255,120),4)
cv2_imshow(a) #(img: MatLike, pt1: Point, pt2: Point, color: Scalar, thickness: int = ..., lineType: int = ..., shift: int = ...)
```

```python
img=np.zeros([200,200,3])
cv2.line(img, (190, 20), (20, 160), (255, 0, 0), 3)
cv2.rectangle(img,(30,30),(170,170),(250,225,120),4, cv2.LINE_8)
cv2.circle(img,(100,100),50,(0,255,0),-1, cv2.LINE_4)
cv2_imshow(img)
```

```python
im=np.zeros([200,200,3])
cv2.rectangle(im,(20,20),(185,185),(250,225,120),4, cv2.LINE_8)
cv2.putText(im,"Hello",(20,90),cv2.FONT_HERSHEY_DUPLEX,2,(0,0,255))
cv2.putText(im,"World",(15,130),cv2.FONT_HERSHEY_DUPLEX,2,(0,0,255))
cv2_imshow(im)  #(img: MatLike, text: str, org: Point, fontFace: int, fontScale: float, color: Scalar, thickness: int = ..., lineType: int = ..., bottomLeftOrigin: bool = ...) -> MatLike
```

```python
from textwrap import fill
imm=np.zeros([300,400,3])
pts = np.array([[50,50],[50,100],[20,200],[250,100],[270,50]])
cv2.polylines(imm, [pts], True, (255,120,120),3)
cv2_imshow(imm) #(img: MatLike, pts: Sequence[MatLike], isClosed: bool, color: Scalar, thickness: int = ..., lineType: int = ..., shift: int = ...)
```

```python
#converting image from colored to binary
import cv2
a= cv2.imread("/content/demo3.png")
ret,binary= cv2.threshold(a,160,255,cv2.THRESH_BINARY)
cv2_imshow(binary)
```
